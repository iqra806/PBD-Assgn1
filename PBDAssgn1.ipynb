```python
#Question 2
#function for finding prime or not
def Prime(value):
    if(value<=1):
        return 0
        
    
    for i in range(2,value):
        if(value%i==0):
            flag=0
            return 0
        
    return 1


#to find consecutive primes
def consecPrimes(val):
    CP=list()
    l=len(val)
    for i in range(l,2,-1):
        j=0
        
        for j in range(0,l):
           
            if((j+(i-1))<=l):
               
                CP.append(val[j:j+i-1])
            else:
                break
    
    return CP


#driven code
n=1
l=1000
primes=list()
maxx=0
p=0;

for i in range(n+1,l):
    if(Prime(i)):
        primes.append(i)

CP=consecPrimes(primes)
for i in range(1,len(CP)-1):
    
    add=int(sum(CP[i]))
    if(Prime(add)):
        if add in primes:
            length=len(CP[i])
            if(length>maxx):
                p=add
                maxx=length
            
              
            
                    
print("sum of the most consecutive primes ",p)
```
